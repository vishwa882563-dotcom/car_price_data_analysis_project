```python
columns=["year","condition","odometer","mmr","sellingprice"]
```

```python
df["year"]=pd.to_numeric(df["year"],errors="coerce")
```

```python
df.drop(["vin","seller","saledate"], axis=1,inplace=True)
```

```python
df.dropna(subset="sellingprice",inplace=True)
```

```python
df["odometer"].fillna(df["odometer"].median(),inplace=True)
df["condition"].fillna(df["condition"].median(),inplace=True)
df["mmr"].fillna(df["mmr"].median(),inplace=True)
```

```python
df["make"].fillna("unknown",inplace=True)
df["model"].fillna("unknown",inplace=True)
df["body"].fillna(df["body"].mode()[0],inplace=True)
df["transmission"].fillna(df["transmission"].mode()[0],inplace=True)
df["trim"].fillna("unknown", inplace=True)
df["color"].fillna("unknown",inplace=True)
```

```python
df.isnull().sum()
```

```python
df["year"]=df["year"].astype(int)
df["condition"]=df["condition"].astype(float)
df["odometer"]=df["odometer"].astype(float)
df["sellingprice"]=df["sellingprice"].astype(float)
```

```python
def remove_outliers(col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return df[(df[col] >= lower) & (df[col] <= upper)]

for col in ['sellingprice','odometer','mmr']:
    df = remove_outliers(col)
```

## data visulation

```python
# price  distribution
plt.figure(figsize=(8,5))
sns.histplot(df["sellingprice"],bins=50,kde=True)
plt.title("price distribution")
plt.show()
```

```python
# avg price by make (top 10)
top_make=df.groupby("make")["sellingprice"].mean().sort_values(ascending=False)

plt.figure(figsize=(15,5))
top_make.plot(kind="bar")
plt.title("Average Selling Price by Make")
plt.show()
```

```python
#  Odometer  vs Price
plt.figure(figsize=(8,5))
sns.scatterplot(x="odometer",y="sellingprice",data=df)
plt.title("Odemeter vs Selling Price")
plt.show()
```

```python
# Price vs Year
plt.figure(figsize=(8,5))
sns.lineplot(x="year",y="sellingprice",data=df)
plt.title("Price Trend by Year")
plt.show()
```
